```python
# importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# reading file and formatting data values
gas_prices = pd.read_csv('U.S._All_Grades_All_Formulations_Retail_Gasoline_Prices.csv')
gas_prices['Dollars per Gallon'] = round(gas_prices['Dollars per Gallon'],2)

gas_prices.head()
```

```python
# plotting gas prices
plt.figure(figsize=(8,4),dpi=300)
plt.xlim(min(gas_prices['Year']),max(gas_prices['Year']))
plt.plot(gas_prices['Year'],gas_prices['Dollars per Gallon'])
plt.show()
```

```python
# reading file
gas_world = pd.read_csv('gas_prices.csv')
gas_world.head()
```

```python
gas_world.tail()
```

```python
# replacing null values with numpy nan
gas_world.replace('N/A',np.nan, inplace=True)
gas_world.head()
```

```python
# creating figure
plt.figure(figsize=(12,7),dpi=400)

# converting columns to strings (for country names) and converting years to integers afterwards
gas_world.columns = [str(col) for col in gas_world.columns]
gas_world['Year'] = gas_world['Year'].astype(int).astype(str)

# dynamically plotting gas prices by country
for country in gas_world.columns:
    if country != 'Year':
        plt.plot(gas_world['Year'],gas_world[country],label=country)

# finalizing visualization
plt.title('Gas Prices by Country Over Time ($)')
plt.xlabel('Year')
plt.ylabel('$ per Gallon')
plt.legend(edgecolor='black')
plt.show()
```
