```python
# replacing state names with abbreviations
for state_name in us_states:
    obesity_data['NAME'].replace(to_replace = state_name, value = us_states[state_name], inplace=True)

obesity_data
```

```python
obesity_data['OBJECTID'] = obesity_data['OBJECTID'].astype(int)
```

```python
obesity_data.head()
```

```python
# creating dataframe from dictionary
state_name_df = pd.DataFrame(list(us_states.items()),columns = ['state name','NAME'])
state_name_df = state_name_df.sort_values(by='NAME',ascending=True)
state_name_df
```

```python
# merge obesity and states dataframes
obesity_data = pd.merge(obesity_data,state_name_df, on = 'NAME',how = 'left')
obesity_data
```

```python
#renaming
obesity_data.rename(columns = {'NAME':'State Abr'},inplace=True)
```

```python
obesity_data.head()
```

```python
state = state_name_df['state name']
state
```

```python
# inserting state column
obesity_data.insert(1,'State Name',state)
obesity_data.head()
```

```python
obesity_data.drop(columns=['state name'],inplace=True)
```

```python
obesity_data = obesity_data.fillna('N/A')
obesity_data = pd.DataFrame(obesity_data)
obesity_data = obesity_data.sort_values(by='Obesity', ascending=False)
obesity_data
```

```python
# obesity greater than 30 = True
high_obesity = obesity_data[obesity_data['Obesity']>=30]
high_obesity
```

```python
# percent of high obesity
high_obesity_percent = round(high_obesity.shape[0]/obesity_data.shape[0],2)*100
high_obesity_percent
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(14,8))
plt.bar(high_obesity['State Abr'],height=high_obesity['Obesity'],width=0.5,edgecolor='black')
plt.ylim(bottom=high_obesity['Obesity'].min()-1)
plt.ylim(top=high_obesity['Obesity'].max()+.5)
plt.title('U.S. States with 30%+ Obesity Rate')
plt.xlabel('State')
plt.ylabel('Percent of Obese Residents')
plt.show()
```
