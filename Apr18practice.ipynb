```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# warm-up: basic graph & adding elements
x=[0,1,2,3,4]
y=[0,2,4,6,8]

plt.plot(x,y,label='2x',color='g',linewidth=2.5,marker='o',markersize=8,markeredgecolor='black',linestyle='--')

plt.title('Basic Graph',fontdict={'fontname':'Times New Roman'},fontsize=15)
plt.xlabel('X',fontdict={'fontname':'Times New Roman'})
plt.ylabel('Y',fontdict={'fontname':'Times New Roman'})

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])
plt.legend()
plt.show()
```

```python
# short hand notation exercise
x=[0,1,2,3,4]
y=[0,2,4,6,8]

plt.figure(figsize=(5,3),dpi=200)
plt.plot(x,y,'go--',label='2x') #short-hand notation

#line 2
x2 = np.arange(0,4,0.5)
print('np.arrangement:',x2)

# plotting both
plt.plot(x2,x2**2,label='x2')
plt.legend(loc='upper left',edgecolor='black')
plt.show()

# saving graph
plt.savefig('apr18.png',dpi=200)
```

```python
# bar chart warm-up
labels = ['A','B','C']
values = [1,4,2]
bars = plt.bar(labels,values)

# creating patterns
patterns = ['X','-','*'] 
for bar in bars:
    bar.set_hatch(patterns.pop(0))

# cleaning visual display
plt.figure(figsize=(7,3),dpi=150)
plt.show()
```
