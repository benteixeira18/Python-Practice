```python
# importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# reading in file and turning it into a dataframe
fifa_data = pd.read_csv('fifa_data.csv')
fifa_data = pd.DataFrame(fifa_data)
fifa_data
```

```python
# creating country-specific data frames
argentina = fifa_data[fifa_data['Nationality']=='Argentina']['Overall']
portugal = fifa_data[fifa_data['Nationality']=='Portugal']['Overall']
brazil = fifa_data[fifa_data['Nationality']=='Brazil']['Overall']
scotland = fifa_data[fifa_data['Nationality']=='Scotland']['Overall']
poland = fifa_data[fifa_data['Nationality']=='Poland']['Overall']

countries = [argentina,portugal,brazil,scotland,poland]
country_labels = ['Argentina','Portugal','Brazil','Scotland','Poland']
```

```python
# box plot of fifa rating by country
plt.figure(figsize=(6,8),dpi=150)
boxes = plt.boxplot(countries,labels=country_labels,patch_artist=True) # patch artist allows you to change facecolor for box plots

# styling boxes for each box plot in graph - edge and fill color
for box in boxes['boxes']:
    box.set(color='#4286f4',linewidth=2)
    box.set(facecolor='#e0e0e0')


# finishing table
plt.title('FIFA Rating by Country')
plt.xlabel('Country')
plt.ylabel('Rating')
plt.grid(visible=True)
plt.show()
```

```python
# histograms - set bins by rating group
bins_ratings = [40,50,60,70,80,90,100]

# plot histogram and line up xticks
plt.figure(figsize=(4,3),dpi=200)
plt.hist(fifa_data.Overall,bins=bins_ratings,width=8,color='#005391')
plt.xticks(bins_ratings)

plt.xlabel('Rating')
plt.ylabel('Player Count')
plt.title('Fifa Player Rating Distribution')
plt.show()
```

```python
# pie chart - preferred foot

# counting number of players preferring left vs. right foot
left = fifa_data[fifa_data['Preferred Foot']=='Left'].count()[0]
right = fifa_data[fifa_data['Preferred Foot']=='Right'].count()[0]
print('Left:',left,'Right:',right)
```

```python
# pie chart of left vs. right with added labeling
total = left + right
percent_left = f"Left: {round(left/total,2)*100}%"
percent_right = f"Right: {round(right/total,2)*100}%"

# plot with labels and colors
plt.pie([left,right],labels=[percent_left,percent_right],colors=['#005391','#FFD700'])
plt.show()
```

```python
# alternative method - autopercentage up to 2 decimals
plt.pie([left,right],labels=['Left','Right'],colors=['#005391','#FFD700'],autopct='%.2f%%')
plt.show()
```

```python
# pie chart of weight

# converting weight to integer
fifa_data.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa_data.Weight]
```

```python
fifa_data.Weight.head()
```

```python
light = fifa_data[fifa_data['Weight']<140].count()[0]
light_medium = fifa_data[(fifa_data['Weight']>=140) & (fifa_data['Weight'] < 160)].count()[0]
medium = fifa_data[(fifa_data['Weight']>=160) & (fifa_data['Weight'] < 180)].count()[0]
medium_heavy = fifa_data[(fifa_data['Weight']>=180) & (fifa_data['Weight'] < 200)].count()[0]
heavy = fifa_data[fifa_data['Weight']>=200].count()[0]
```

```python
# labeling
weights = [light,light_medium,medium,medium_heavy,heavy]
labels = ['featherweight','light weight','middle weight','light heavyweight','heavyweight']

# plotting, using ggplot color style, and formatting to make data read clearer
plt.figure(figsize=(4,4),dpi=500)
plt.style.use('ggplot')
plt.pie(weights,labels=labels,autopct='%.0f%%',pctdistance=0.7,explode = (0.1, 0, 0, 0,0.1)) #explode highlights outliers better

# finalizing
plt.title('Weights of FIFA Players by Weight Class (lbs)')
plt.show()
```
