```python
# importing libraries
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np

# setting option to display all rows and columns
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
```

```python
# getting nba data
bos_gsw = pd.read_csv('warriorsceltics.csv')
bos_gsw = pd.DataFrame(bos_gsw)
bos_gsw
```

```python
# formatting dates
bos_gsw['Date'] = pd.to_datetime(bos_gsw['Date'])
bos_gsw
```

```python
# cleaning dates 

# previewing wrong years - 1974 and earlier
bos_gsw[bos_gsw['Date'].dt.year >2025] 
```

```python
# changing dates by offsetting the years of "2074" (1974) and earlier
from pandas.tseries.offsets import DateOffset # library used to offset dates by years
```

```python
# Offsetting date years of 2046+ by subtraction of 100 (not changing any other dates, datatypes or columns)
bos_gsw.loc[bos_gsw['Date'].dt.year >= 2046,'Date'] -= DateOffset(years=100)
bos_gsw.tail()
```

```python
# replacng null values
bos_gsw['Home Team'].fillna(np.nan)
bos_gsw['Away Team'].fillna(np.nan)
```

```python
# comparing and plotting Celtics vs. Warriors Wins

# counting wins by franchise
bos_wins = bos_gsw[bos_gsw['Win or Loss for Celtics']=='W'].count()[0]
gsw_wins = bos_gsw[bos_gsw['Win or Loss for Warriors']=='W'].count()[0]

# plotting wins
plt.figure(figsize=(5,4),dpi=200)
plt.bar('Celtics',bos_wins,color='#007A33')
plt.bar('Warriors',gsw_wins,color = '#FFC72C')

# labeling bar chart
plt.title('Cetlics vs. Warriors Wins Over Time (1946-2022)')
plt.ylabel('Wins')
plt.show()
```

```python
# plotting points scored over time
bos_points = bos_gsw['Celtics score']
gsw_points = bos_gsw['Warriors score']

plt.figure(figsize=(16,10),dpi=300)
plt.plot(bos_gsw['Date'].dt.year,bos_points,color='#007A33')
plt.plot(bos_gsw['Date'].dt.year,gsw_points,color = '#FFC72C')

# final labels
plt.title('Points Scored per Game Over Time Between Cs & Ws Matchups')
plt.xlabel('Year Game Played In')
plt.ylabel('Points Scored Per Game')
plt.legend(labels = ['Celtics','Warriors'],fontsize='xx-large')
plt.show()
```

```python
# showing point differential between Celtics and Warriors
bos_points_diff = bos_gsw['Celtics score']-bos_gsw['Warriors score']
plt.figure(figsize=(16,8),dpi=200)
bins = [-50,-40,-30,-20,-10,0,10,20,30,40,50]
plt.hist(bos_points_diff,color='#007A33',bins = bins,align='left',rwidth=0.8,edgecolor='black') # number of games

# labeling
plt.show()
```

```python
# pie chart showing comparison of point differential

# counting and categorizing wins
blowout_win = bos_points_diff[bos_points_diff >= 30].count()
great_win = bos_points_diff[(bos_points_diff >=20)&(bos_points_diff < 30)].count()
good_win = bos_points_diff[(bos_points_diff >=10)&(bos_points_diff < 20)].count()
close_win = bos_points_diff[(bos_points_diff >0)&(bos_points_diff < 10)].count()

# counting and categorizing losses
close_loss = bos_points_diff[(bos_points_diff >-10)&(bos_points_diff < 0)].count()
bad_loss = bos_points_diff[(bos_points_diff >-20)&(bos_points_diff < -10)].count()
awful_loss = bos_points_diff[(bos_points_diff >-30)&(bos_points_diff < -20)].count()
blowout_loss = bos_points_diff[bos_points_diff <= -30].count()
```

```python
# creating labels for pie chart
points_categories = [blowout_win,great_win,good_win,close_win,close_loss,bad_loss,awful_loss,blowout_loss]
points_labels = ['Blowout Win','Great Win','Good Win','Close Win','Close Loss','Bad Loss','Awful Loss','Blowout Loss']

# plotting pie chart
plt.figure(figsize = (3,2),dpi=200)
explode = (0.3,0,0,0,0,0,0.3,0.3) # sets apart smaller categories to increase visibility
plt.pie(points_categories,explode=explode,labels=points_labels,autopct = '%.0f%%',radius=2.5,
        colors = ['#007A33','#007A33','#007A33','grey','grey','red','red','red'],pctdistance=0.8)
plt.title('Outcomes of Celtics games vs. Warriors',y=1.75)
plt.show()
```
